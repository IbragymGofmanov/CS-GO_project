```python
from PyQt6 import QtCore, QtGui, QtWidgets
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_qt5agg import FigureCanvasQTAgg as FigureCanvas
from matplotlib.figure import Figure

class Ui_mmCSGO(object):
    def setupUi(self, mmCSGO):
        mmCSGO.setObjectName("mmCSGO")
        mmCSGO.resize(654, 619)
        self.gridLayout = QtWidgets.QGridLayout(mmCSGO)
        self.gridLayout.setObjectName("gridLayout")
        mmCSGO.setStyleSheet('background-image: url(gradient.jpg);')

        # ====================== STACKED WIDGET (основной контейнер страниц) ======================
        self.stackedWidget = QtWidgets.QStackedWidget(parent=mmCSGO)
        sizePolicy = QtWidgets.QSizePolicy(QtWidgets.QSizePolicy.Policy.Preferred, QtWidgets.QSizePolicy.Policy.Preferred)
        sizePolicy.setHorizontalStretch(0)
        sizePolicy.setVerticalStretch(0)
        sizePolicy.setHeightForWidth(self.stackedWidget.sizePolicy().hasHeightForWidth())
        self.stackedWidget.setSizePolicy(sizePolicy)
        self.stackedWidget.setObjectName("stackedWidget")

        # ====================== СТРАНИЦА 0: MAIN (Главное меню) ======================
        self.Main = QtWidgets.QWidget()
        self.Main.setObjectName("Main")
        self.gridLayout_3 = QtWidgets.QGridLayout(self.Main)

        self.gridLayout_3.setObjectName("gridLayout_3")
        self.gridLayout_2 = QtWidgets.QGridLayout()
        self.gridLayout_2.setObjectName("gridLayout_2")

        # --- Кнопка "ИНФО" (переход на страницу 3 - Info) ---
        self.Main_pushButton_Info = QtWidgets.QPushButton(parent=self.Main)
        font = QtGui.QFont()
        font.setFamily("Counter-Strike")
        font.setPointSize(30)
        self.Main_pushButton_Info.setFont(font)
        self.Main_pushButton_Info.setCursor(QtGui.QCursor(QtCore.Qt.CursorShape.CrossCursor))
        self.Main_pushButton_Info.setObjectName("Main_pushButton_Info")
        self.Main_pushButton_Info.clicked.connect(lambda: self.go_to_info(0))
        self.Main_pushButton_Info.setStyleSheet("""
        QPushButton {background: transparent;
        color: black;  /* Черный цвет текста */
        border: none;  /* Убираем границу */
        padding: 0px;  /* Убираем отступы */}
        QPushButton:hover {color: #333;  /* Темно-серый при наведении */}QPushButton:pressed {color: #555;  /* Серый при нажатии */}""")

        self.gridLayout_2.addWidget(self.Main_pushButton_Info, 2, 1, 1, 1)

        # --- Кнопка "НАЧАТЬ" (переход на страницу 1 - Plan) ---
        self.Main_pushButton_Start = QtWidgets.QPushButton(parent=self.Main)
        sizePolicy = QtWidgets.QSizePolicy(QtWidgets.QSizePolicy.Policy.Minimum, QtWidgets.QSizePolicy.Policy.Fixed)
        sizePolicy.setHorizontalStretch(0)
        sizePolicy.setVerticalStretch(0)
        sizePolicy.setHeightForWidth(self.Main_pushButton_Start.sizePolicy().hasHeightForWidth())
        self.Main_pushButton_Start.setSizePolicy(sizePolicy)
        font = QtGui.QFont()
        font.setFamily("Counter-Strike")
        font.setPointSize(30)
        self.Main_pushButton_Start.setFont(font)
        self.Main_pushButton_Start.setCursor(QtGui.QCursor(QtCore.Qt.CursorShape.CrossCursor))
        self.Main_pushButton_Start.setStyleSheet("background: transparent")
        self.Main_pushButton_Start.setObjectName("Main_pushButton_Start")
        self.Main_pushButton_Start.clicked.connect(lambda: self.stackedWidget.setCurrentIndex(1))
        self.Main_pushButton_Start.setStyleSheet("""
        QPushButton {background: transparent;
        color: black;  /* Черный цвет текста */
        border: none;  /* Убираем границу */
        padding: 0px;  /* Убираем отступы */}
        QPushButton:hover {color: #333;  /* Темно-серый при наведении */}QPushButton:pressed {color: #555;  /* Серый при нажатии */}""")

        self.gridLayout_2.addWidget(self.Main_pushButton_Start, 1, 1, 1, 1)

        # --- Кнопка "ВЫХОД" (закрытие приложения) ---
        self.Main_pushButton_Exit = QtWidgets.QPushButton(parent=self.Main)
        font = QtGui.QFont()
        font.setFamily("Counter-Strike")
        font.setPointSize(30)
        self.Main_pushButton_Exit.setFont(font)
        self.Main_pushButton_Exit.setCursor(QtGui.QCursor(QtCore.Qt.CursorShape.CrossCursor))
        self.Main_pushButton_Exit.setStyleSheet("background: transparent")
        self.Main_pushButton_Exit.setObjectName("Main_pushButton_Exit")
        self.Main_pushButton_Exit.clicked.connect(mmCSGO.close)
        self.Main_pushButton_Exit.setStyleSheet("""
        QPushButton {background: transparent;
        color: black;  /* Черный цвет текста */
        border: none;  /* Убираем границу */
        padding: 0px;  /* Убираем отступы */}
        QPushButton:hover {color: #333;  /* Темно-серый при наведении */}QPushButton:pressed {color: #555;  /* Серый при нажатии */}""")

        self.gridLayout_2.addWidget(self.Main_pushButton_Exit, 3, 1, 1, 1)

        # --- Заголовок ---
        self.Main_Label_mmCSGO = QtWidgets.QLabel(parent=self.Main)
        self.Main_Label_mmCSGO.setMaximumSize(QtCore.QSize(16777215, 200))
        font = QtGui.QFont()
        font.setFamily("Counter-Strike")
        font.setPointSize(64)
        self.Main_Label_mmCSGO.setFont(font)
        self.Main_Label_mmCSGO.setCursor(QtGui.QCursor(QtCore.Qt.CursorShape.ArrowCursor))
        self.Main_Label_mmCSGO.setAlignment(QtCore.Qt.AlignmentFlag.AlignCenter)
        self.Main_Label_mmCSGO.setObjectName("Main_Label_mmCSGO")
        self.Main_Label_mmCSGO.setStyleSheet("""QLabel {background: transparent; color: black;}""")

        self.gridLayout_2.addWidget(self.Main_Label_mmCSGO, 0, 1, 1, 1)
        spacerItem = QtWidgets.QSpacerItem(40, 20, QtWidgets.QSizePolicy.Policy.Preferred, QtWidgets.QSizePolicy.Policy.Minimum)
        self.gridLayout_2.addItem(spacerItem, 1, 0, 1, 1)
        spacerItem1 = QtWidgets.QSpacerItem(40, 20, QtWidgets.QSizePolicy.Policy.Preferred, QtWidgets.QSizePolicy.Policy.Minimum)
        self.gridLayout_2.addItem(spacerItem1, 1, 2, 1, 1)
        self.gridLayout_3.addLayout(self.gridLayout_2, 0, 0, 1, 1)
        self.stackedWidget.addWidget(self.Main)

        # ====================== СТРАНИЦА 1: PLAN (План задачи) ======================
        self.Plan = QtWidgets.QWidget()
        self.Plan.setObjectName("Plan")
        self.gridLayout_4 = QtWidgets.QGridLayout(self.Plan)
        self.gridLayout_4.setObjectName("gridLayout_4")
        self.verticalLayout_3 = QtWidgets.QVBoxLayout()
        self.verticalLayout_3.setObjectName("verticalLayout_3")
        self.horizontalLayout_4 = QtWidgets.QHBoxLayout()
        self.horizontalLayout_4.setObjectName("horizontalLayout_4")
        self.verticalLayout_2 = QtWidgets.QVBoxLayout()
        self.verticalLayout_2.setObjectName("verticalLayout_2")

        # --- Лейбл "ПЛАН ЗАДАЧИ" ---
        self.Plan_Label_Plan = QtWidgets.QLabel(parent=self.Plan)
        self.Plan_Label_Plan.setObjectName("Plan_Label_Plan")
        font = QtGui.QFont()
        font.setFamily("Counter-Strike")
        font.setPointSize(48)
        self.Plan_Label_Plan.setFont(font)
        self.Plan_Label_Plan.setStyleSheet("""QLabel {background: transparent; color: black;}""")
        self.verticalLayout_2.addWidget(self.Plan_Label_Plan)

        # --- Лейбл "Параметры" ---
        self.Plan_Label_Parameters = QtWidgets.QLabel(parent=self.Plan)
        self.Plan_Label_Parameters.setObjectName("Plan_Label_Parameters")
        font = QtGui.QFont()
        font.setFamily("Counter-Strike")
        font.setPointSize(30)
        self.Plan_Label_Parameters.setFont(font)
        self.Plan_Label_Parameters.setStyleSheet("""QLabel {background: transparent; color: black;}""")
        self.verticalLayout_2.addWidget(self.Plan_Label_Parameters)
        self.Plan_lineEdit_a = QtWidgets.QLineEdit(parent=self.Plan)
        self.Plan_lineEdit_a.setObjectName("Plan_lineEdit_a")
        self.verticalLayout_2.addWidget(self.Plan_lineEdit_a)
        self.Plan_lineEdit_p = QtWidgets.QLineEdit(parent=self.Plan)
        self.Plan_lineEdit_p.setObjectName("Plan_lineEdit_p")
        self.verticalLayout_2.addWidget(self.Plan_lineEdit_p)
        self.Plan_lineEdit_q = QtWidgets.QLineEdit(parent=self.Plan)
        self.Plan_lineEdit_q.setObjectName("Plan_lineEdit_q")
        self.verticalLayout_2.addWidget(self.Plan_lineEdit_q)

        # --- Лейбл "ОДЗ" ---
        self.Plan_Label_Problem = QtWidgets.QLabel(parent=self.Plan)
        self.Plan_Label_Problem.setObjectName("Plan_Label_Problem")
        font = QtGui.QFont()
        font.setFamily("Counter-Strike")
        font.setPointSize(16)
        self.Plan_Label_Problem.setFont(font)
        self.Plan_Label_Problem.setStyleSheet("""QLabel {background: transparent; color: black;}""")
        self.verticalLayout_2.addWidget(self.Plan_Label_Problem)
        self.horizontalLayout_5 = QtWidgets.QHBoxLayout()
        self.horizontalLayout_5.setObjectName("horizontalLayout_5")

        # --- Кнопкa "Стандартные значения" (ждём Ярика) ---
        self.Plan_pushButton_Standart = QtWidgets.QPushButton(parent=self.Plan)
        self.Plan_pushButton_Standart.setObjectName("Plan_pushButton_Standart")
        self.Plan_pushButton_Standart.setCursor(QtGui.QCursor(QtCore.Qt.CursorShape.ArrowCursor))
        self.horizontalLayout_5.addWidget(self.Plan_pushButton_Standart)
        font = QtGui.QFont()
        font.setFamily("Counter-Strike")
        font.setPointSize(18)
        self.Plan_pushButton_Standart.setFont(font)
        self.Plan_pushButton_Standart.setCursor(QtGui.QCursor(QtCore.Qt.CursorShape.CrossCursor))
        self.Plan_pushButton_Standart.setStyleSheet("""
        QPushButton {background: transparent;
        color: black;  /* Черный цвет текста */
        border: none;  /* Убираем границу */
        padding: 0px;  /* Убираем отступы */}
        QPushButton:hover {color: #333;  /* Темно-серый при наведении */}QPushButton:pressed {color: #555;  /* Серый при нажатии */}""")

        # --- Кнопка "Рандомные значения" (ждём Ярика) ---
        self.Plan_pushButton_Random = QtWidgets.QPushButton(parent=self.Plan)
        self.Plan_pushButton_Random.setObjectName("Plan_pushButton_Random")
        self.Plan_pushButton_Random.setCursor(QtGui.QCursor(QtCore.Qt.CursorShape.ArrowCursor))
        font = QtGui.QFont()
        font.setFamily("Counter-Strike")
        font.setPointSize(18)
        self.Plan_pushButton_Random.setFont(font)
        self.Plan_pushButton_Random.setCursor(QtGui.QCursor(QtCore.Qt.CursorShape.CrossCursor))
        self.Plan_pushButton_Random.setStyleSheet("""
        QPushButton {background: transparent;
        color: black;  /* Черный цвет текста */
        border: none;  /* Убираем границу */
        padding: 0px;  /* Убираем отступы */}
        QPushButton:hover {color: #333;  /* Темно-серый при наведении */}QPushButton:pressed {color: #555;  /* Серый при нажатии */}""")

        self.horizontalLayout_5.addWidget(self.Plan_pushButton_Random)
        self.verticalLayout_2.addLayout(self.horizontalLayout_5)
        self.horizontalLayout_4.addLayout(self.verticalLayout_2)
        self.plot_widget1 = self.create_plot_widget()
        self.horizontalLayout_4.addWidget(self.plot_widget1)

        self.verticalLayout_3.addLayout(self.horizontalLayout_4)
        self.gridLayout_8 = QtWidgets.QGridLayout()
        self.gridLayout_8.setObjectName("gridLayout_8")
        spacerItem3 = QtWidgets.QSpacerItem(40, 20, QtWidgets.QSizePolicy.Policy.Expanding, QtWidgets.QSizePolicy.Policy.Minimum)
        self.gridLayout_8.addItem(spacerItem3, 1, 0, 1, 1)
        spacerItem4 = QtWidgets.QSpacerItem(40, 20, QtWidgets.QSizePolicy.Policy.Expanding, QtWidgets.QSizePolicy.Policy.Minimum)
        self.gridLayout_8.addItem(spacerItem4, 1, 2, 1, 1)
        self.Plan_pushButton_Standart.setStyleSheet("""
        QPushButton {background: transparent;
        color: black;  /* Черный цвет текста */
        border: none;  /* Убираем границу */
        padding: 0px;  /* Убираем отступы */}
        QPushButton:hover {color: #333;  /* Темно-серый при наведении */}QPushButton:pressed {color: #555;  /* Серый при нажатии */}""")
        self.Plan_pushButton_Random.setStyleSheet("""
        QPushButton {background: transparent;
        color: black;  /* Черный цвет текста */
        border: none;  /* Убираем границу */
        padding: 0px;  /* Убираем отступы */}
        QPushButton:hover {color: #333;  /* Темно-серый при наведении */}QPushButton:pressed {color: #555;  /* Серый при нажатии */}""")

        # --- Кнопка "ИНФО" (переход на страницу 3 - Info) ---
        self.Plan_pushButton_Info = QtWidgets.QPushButton(parent=self.Plan)
        self.Plan_pushButton_Info.setObjectName("Plan_pushButton_Info")
        self.Plan_pushButton_Info.clicked.connect(lambda: self.go_to_info(1))
        self.Plan_pushButton_Info.setCursor(QtGui.QCursor(QtCore.Qt.CursorShape.CrossCursor))
        font = QtGui.QFont()
        font.setFamily("Counter-Strike")
        font.setPointSize(24)
        self.Plan_pushButton_Info.setFont(font)
        self.Plan_pushButton_Info.setStyleSheet("""
        QPushButton {background: transparent;
        color: black;  /* Черный цвет текста */
        border: none;  /* Убираем границу */
        padding: 0px;  /* Убираем отступы */}
        QPushButton:hover {color: #333;  /* Темно-серый при наведении */}QPushButton:pressed {color: #555;  /* Серый при нажатии */}""")

        self.gridLayout_8.addWidget(self.Plan_pushButton_Info, 1, 1, 1, 1)
        spacerItem5 = QtWidgets.QSpacerItem(40, 20, QtWidgets.QSizePolicy.Policy.Expanding, QtWidgets.QSizePolicy.Policy.Minimum)
        self.gridLayout_8.addItem(spacerItem5, 0, 1, 1, 1)

        # --- Кнопка "НАЗАД" (возврат на Main) ---
        self.Plan_pushButton_Back = QtWidgets.QPushButton(parent=self.Plan)
        self.Plan_pushButton_Back.setObjectName("Plan_pushButton_Back")
        self.Plan_pushButton_Back.clicked.connect(lambda: self.stackedWidget.setCurrentIndex(0))
        self.Plan_pushButton_Back.setCursor(QtGui.QCursor(QtCore.Qt.CursorShape.CrossCursor))
        font = QtGui.QFont()
        font.setFamily("Counter-Strike")
        font.setPointSize(24)
        self.Plan_pushButton_Back.setFont(font)
        self.Plan_pushButton_Back.setStyleSheet("""
        QPushButton {background: transparent;
        color: black;  /* Черный цвет текста */
        border: none;  /* Убираем границу */
        padding: 0px;  /* Убираем отступы */}
        QPushButton:hover {color: #333;  /* Темно-серый при наведении */}QPushButton:pressed {color: #555;  /* Серый при нажатии */}""")

        self.gridLayout_8.addWidget(self.Plan_pushButton_Back, 0, 0, 1, 1)

        # --- Кнопка "ДАЛЕЕ" (переход на Decision) ---
        self.Plan_pushButton_Next = QtWidgets.QPushButton(parent=self.Plan)
        self.Plan_pushButton_Next.setObjectName("Plan_pushButton_Next")
        self.Plan_pushButton_Next.clicked.connect(lambda: self.stackedWidget.setCurrentIndex(2))
        self.Plan_pushButton_Next.setCursor(QtGui.QCursor(QtCore.Qt.CursorShape.CrossCursor))
        font = QtGui.QFont()
        font.setFamily("Counter-Strike")
        font.setPointSize(24)
        self.Plan_pushButton_Next.setFont(font)
        self.Plan_pushButton_Next.setStyleSheet("""
        QPushButton {background: transparent;
        color: black;  /* Черный цвет текста */
        border: none;  /* Убираем границу */
        padding: 0px;  /* Убираем отступы */}
        QPushButton:hover {color: #333;  /* Темно-серый при наведении */}QPushButton:pressed {color: #555;  /* Серый при нажатии */}""")

        self.gridLayout_8.addWidget(self.Plan_pushButton_Next, 0, 2, 1, 1)
        self.verticalLayout_3.addLayout(self.gridLayout_8)
        self.gridLayout_4.addLayout(self.verticalLayout_3, 0, 0, 1, 1)
        self.stackedWidget.addWidget(self.Plan)

        # ====================== СТРАНИЦА 2: DECISION (Решение задачи) ======================
        self.Decision = QtWidgets.QWidget()
        self.Decision.setObjectName("Decision")
        self.gridLayout_9 = QtWidgets.QGridLayout(self.Decision)
        self.gridLayout_9.setObjectName("gridLayout_9")
        self.Decision_gridLayout_7 = QtWidgets.QGridLayout()
        self.Decision_gridLayout_7.setObjectName("Decision_gridLayout_7")
        self.Desicion_verticalLayout_4 = QtWidgets.QVBoxLayout()
        self.Desicion_verticalLayout_4.setObjectName("Desicion_verticalLayout_4")
        self.Desion_horizontalLayout_6 = QtWidgets.QHBoxLayout()
        self.Desion_horizontalLayout_6.setSizeConstraint(QtWidgets.QLayout.SizeConstraint.SetMaximumSize)
        self.Desion_horizontalLayout_6.setSpacing(6)
        self.Desion_horizontalLayout_6.setObjectName("Desion_horizontalLayout_6")
        self.Desion_verticalLayout_5 = QtWidgets.QVBoxLayout()
        self.Desion_verticalLayout_5.setObjectName("Desion_verticalLayout_5")
        self.Decision_Label_Decision = QtWidgets.QLabel(parent=self.Decision)

        self.Decision_Label_Decision.setObjectName("Decision_Label_Decision")
        self.Desion_verticalLayout_5.addWidget(self.Decision_Label_Decision)
        font = QtGui.QFont()
        font.setFamily("Counter-Strike")
        font.setPointSize(48)
        self.Decision_Label_Decision.setFont(font)
        self.Decision_Label_Decision.setStyleSheet("""QLabel {background: transparent; color: black;}""")

        self.Desion_verticalLayout_6 = QtWidgets.QVBoxLayout()
        self.Desion_verticalLayout_6.setObjectName("Desion_verticalLayout_6")
        self.Decision_Label_Parameters = QtWidgets.QLabel(parent=self.Decision)
        self.Decision_Label_Parameters.setObjectName("Decision_Label_Parameters")
        self.Desion_verticalLayout_6.addWidget(self.Decision_Label_Parameters)
        font = QtGui.QFont()
        font.setFamily("Counter-Strike")
        font.setPointSize(16)
        self.Decision_Label_Parameters.setFont(font)
        self.Decision_Label_Parameters.setStyleSheet("""QLabel {background: transparent; color: black;}""")

        self.Decision_Label_a = QtWidgets.QLabel(parent=self.Decision)

        self.Decision_Label_a.setObjectName("Decision_Label_a")
        self.Desion_verticalLayout_6.addWidget(self.Decision_Label_a)
        font = QtGui.QFont()
        font.setFamily("Counter-Strike")
        font.setPointSize(16)
        self.Decision_Label_a.setFont(font)
        self.Decision_Label_a.setStyleSheet("""QLabel {background: transparent; color: black;}""")

        self.Decision_Label_p = QtWidgets.QLabel(parent=self.Decision)
        self.Decision_Label_p.setObjectName("Decision_Label_p")
        self.Desion_verticalLayout_6.addWidget(self.Decision_Label_p)
        font = QtGui.QFont()
        font.setFamily("Counter-Strike")
        font.setPointSize(16)
        self.Decision_Label_p.setFont(font)
        self.Decision_Label_p.setStyleSheet("""QLabel {background: transparent; color: black;}""")

        self.Decision_Label_q = QtWidgets.QLabel(parent=self.Decision)
        self.Decision_Label_q.setObjectName("Decision_Label_q")
        self.Desion_verticalLayout_6.addWidget(self.Decision_Label_q)
        font = QtGui.QFont()
        font.setFamily("Counter-Strike")
        font.setPointSize(16)
        self.Decision_Label_q.setFont(font)
        self.Decision_Label_q.setStyleSheet("""QLabel {background: transparent; color: black;}""")

        self.Desion_verticalLayout_5.addLayout(self.Desion_verticalLayout_6)

        self.Decision_Label_Task = QtWidgets.QLabel(parent=self.Decision)
        self.Decision_Label_Task.setObjectName("Decision_Label_Task")
        self.Desion_verticalLayout_5.addWidget(self.Decision_Label_Task)
        font = QtGui.QFont()
        font.setFamily("Counter-Strike")
        font.setPointSize(16)
        self.Decision_Label_Task.setFont(font)
        self.Decision_Label_Task.setStyleSheet("""QLabel {background: transparent; color: black;}""")

        self.Decision_textBrowser_Task = QtWidgets.QTextBrowser(parent=self.Decision)
        sizePolicy = QtWidgets.QSizePolicy(QtWidgets.QSizePolicy.Policy.Minimum, QtWidgets.QSizePolicy.Policy.Expanding)
        sizePolicy.setHorizontalStretch(0)
        sizePolicy.setVerticalStretch(0)
        sizePolicy.setHeightForWidth(self.Decision_textBrowser_Task.sizePolicy().hasHeightForWidth())
        self.Decision_textBrowser_Task.setSizePolicy(sizePolicy)
        self.Decision_textBrowser_Task.setMaximumSize(QtCore.QSize(267, 16777215))
        self.Decision_textBrowser_Task.setContextMenuPolicy(QtCore.Qt.ContextMenuPolicy.DefaultContextMenu)
        self.Decision_textBrowser_Task.setStyleSheet("background: transparent")
        self.Decision_textBrowser_Task.setObjectName("Decision_textBrowser_Task")
        self.Desion_verticalLayout_5.addWidget(self.Decision_textBrowser_Task)
        self.Desion_horizontalLayout_6.addLayout(self.Desion_verticalLayout_5)
        self.Desion_verticalLayout_12 = QtWidgets.QVBoxLayout()
        self.Desion_verticalLayout_12.setObjectName("Desion_verticalLayout_12")
        self.plot_widget2 = self.create_plot_widget()
        self.Desion_verticalLayout_12.addWidget(self.plot_widget2)

        self.Desion_formLayout_3 = QtWidgets.QFormLayout()
        self.Desion_formLayout_3.setObjectName("Desion_formLayout_3")
        self.Desion_Label = QtWidgets.QLabel(parent=self.Decision)
        self.Desion_Label.setObjectName("Desion_Label")
        font = QtGui.QFont()
        font.setFamily("Counter-Strike")
        font.setPointSize(16)
        self.Desion_Label.setFont(font)
        self.Desion_Label.setStyleSheet("""QLabel {background: transparent; color: black;}""")
        self.Desion_formLayout_3.setWidget(0, QtWidgets.QFormLayout.ItemRole.LabelRole, self.Desion_Label)

        self.Desion_LineEdit = QtWidgets.QLineEdit(parent=self.Decision)
        self.Desion_LineEdit.setFrame(False)
        self.Desion_LineEdit.setObjectName("Desion_LineEdit")
        self.Desion_formLayout_3.setWidget(0, QtWidgets.QFormLayout.ItemRole.FieldRole, self.Desion_LineEdit)
        self.Desion_verticalLayout_12.addLayout(self.Desion_formLayout_3)
        self.Desion_horizontalLayout_6.addLayout(self.Desion_verticalLayout_12)
        self.Desicion_verticalLayout_4.addLayout(self.Desion_horizontalLayout_6)
        self.Desion_gridLayout_10 = QtWidgets.QGridLayout()
        self.Desion_gridLayout_10.setObjectName("Desion_gridLayout_10")
        self.Decision_pushButton_Back = QtWidgets.QPushButton(parent=self.Decision)
        self.Decision_pushButton_Back.setObjectName("Decision_pushButton_Back")
        self.Decision_pushButton_Back.clicked.connect(lambda: self.stackedWidget.setCurrentIndex(1))
        self.Decision_pushButton_Back.setCursor(QtGui.QCursor(QtCore.Qt.CursorShape.CrossCursor))
        font = QtGui.QFont()
        font.setFamily("Counter-Strike")
        font.setPointSize(24)
        self.Decision_pushButton_Back.setFont(font)
        self.Decision_pushButton_Back.setStyleSheet("""
        QPushButton {background: transparent;
        color: black;  /* Черный цвет текста */
        border: none;  /* Убираем границу */
        padding: 0px;  /* Убираем отступы */}
        QPushButton:hover {color: #333;  /* Темно-серый при наведении */}QPushButton:pressed {color: #555;  /* Серый при нажатии */}""")

        self.Desion_gridLayout_10.addWidget(self.Decision_pushButton_Back, 0, 0, 1, 1)
        self.Decision_pushButton_Exit = QtWidgets.QPushButton(parent=self.Decision)
        self.Decision_pushButton_Exit.setStyleSheet("background: transparent")
        self.Decision_pushButton_Exit.setObjectName("Decision_pushButton_Exit")
        font = QtGui.QFont()
        font.setFamily("Counter-Strike")
        font.setPointSize(30)
        self.Decision_pushButton_Exit.setFont(font)
        self.Decision_pushButton_Exit.setCursor(QtGui.QCursor(QtCore.Qt.CursorShape.CrossCursor))
        self.Decision_pushButton_Exit.clicked.connect(mmCSGO.close)
        self.Decision_pushButton_Exit.setStyleSheet("""
        QPushButton {background: transparent;
        color: black;  /* Черный цвет текста */
        border: none;  /* Убираем границу */
        padding: 0px;  /* Убираем отступы */}
        QPushButton:hover {color: #333;  /* Темно-серый при наведении */}QPushButton:pressed {color: #555;  /* Серый при нажатии */}""")

        self.Desion_gridLayout_10.addWidget(self.Decision_pushButton_Exit, 0, 2, 1, 1)
        spacerItem8 = QtWidgets.QSpacerItem(40, 20, QtWidgets.QSizePolicy.Policy.Expanding, QtWidgets.QSizePolicy.Policy.Minimum)
        self.Desion_gridLayout_10.addItem(spacerItem8, 0, 1, 1, 1)
        self.Decision_pushButton_Info = QtWidgets.QPushButton(parent=self.Decision)
        self.Decision_pushButton_Info.setStyleSheet("background: transparent")
        self.Decision_pushButton_Info.setObjectName("Decision_pushButton_Info")
        self.Decision_pushButton_Info.clicked.connect(lambda: self.go_to_info(1))
        self.Decision_pushButton_Info.setCursor(QtGui.QCursor(QtCore.Qt.CursorShape.CrossCursor))
        font = QtGui.QFont()
        font.setFamily("Counter-Strike")
        font.setPointSize(24)
        self.Decision_pushButton_Info.setFont(font)
        self.Decision_pushButton_Info.setStyleSheet("""
        QPushButton {background: transparent;
        color: black;  /* Черный цвет текста */
        border: none;  /* Убираем границу */
        padding: 0px;  /* Убираем отступы */}
        QPushButton:hover {color: #333;  /* Темно-серый при наведении */}QPushButton:pressed {color: #555;  /* Серый при нажатии */}""")

        self.Desion_gridLayout_10.addWidget(self.Decision_pushButton_Info, 1, 1, 1, 1)
        spacerItem9 = QtWidgets.QSpacerItem(40, 20, QtWidgets.QSizePolicy.Policy.Expanding, QtWidgets.QSizePolicy.Policy.Minimum)
        self.Desion_gridLayout_10.addItem(spacerItem9, 1, 0, 1, 1)
        spacerItem10 = QtWidgets.QSpacerItem(40, 20, QtWidgets.QSizePolicy.Policy.Expanding, QtWidgets.QSizePolicy.Policy.Minimum)
        self.Desion_gridLayout_10.addItem(spacerItem10, 1, 2, 1, 1)
        self.Desicion_verticalLayout_4.addLayout(self.Desion_gridLayout_10)
        self.Decision_gridLayout_7.addLayout(self.Desicion_verticalLayout_4, 0, 0, 1, 1)
        self.gridLayout_9.addLayout(self.Decision_gridLayout_7, 0, 0, 1, 1)
        self.stackedWidget.addWidget(self.Decision)

        # ====================== СТРАНИЦА 3: INFO (Справочник) ======================
        self.Info = QtWidgets.QWidget()
        self.Info.setObjectName("Info")
        self.gridLayout_12 = QtWidgets.QGridLayout(self.Info)
        self.gridLayout_12.setObjectName("gridLayout_12")
        self.gridLayout_11 = QtWidgets.QGridLayout()
        self.gridLayout_11.setObjectName("gridLayout_11")
        self.verticalLayout_7 = QtWidgets.QVBoxLayout()
        self.verticalLayout_7.setObjectName("verticalLayout_7")
        self.horizontalLayout_9 = QtWidgets.QHBoxLayout()
        self.horizontalLayout_9.setObjectName("horizontalLayout_9")
        spacerItem10 = QtWidgets.QSpacerItem(40, 20, QtWidgets.QSizePolicy.Policy.Expanding, QtWidgets.QSizePolicy.Policy.Minimum)
        self.horizontalLayout_9.addItem(spacerItem10)
        self.Info_Label_Info = QtWidgets.QLabel(parent=self.Info)
        self.Info_Label_Info.setObjectName("Info_Label_Info")
        font.setFamily("Counter-Strike")
        font.setPointSize(48)
        self.Info_Label_Info.setFont(font)
        self.Info_Label_Info.setStyleSheet("""QLabel {background: transparent; color: black;}""")
        self.horizontalLayout_9.addWidget(self.Info_Label_Info)
        spacerItem11 = QtWidgets.QSpacerItem(40, 20, QtWidgets.QSizePolicy.Policy.Expanding, QtWidgets.QSizePolicy.Policy.Minimum)
        self.horizontalLayout_9.addItem(spacerItem11)
        self.verticalLayout_7.addLayout(self.horizontalLayout_9)
        self.horizontalLayout_7 = QtWidgets.QHBoxLayout()
        self.horizontalLayout_7.setObjectName("horizontalLayout_7")
        self.verticalLayout_8 = QtWidgets.QVBoxLayout()
        self.verticalLayout_8.setObjectName("verticalLayout_8")
        self.verticalLayout_10 = QtWidgets.QVBoxLayout()
        self.verticalLayout_10.setObjectName("verticalLayout_10")
        self.Info_Label_Authors = QtWidgets.QLabel(parent=self.Info)
        self.Info_Label_Authors.setObjectName("Info_Label_Authors")
        font.setFamily("Counter-Strike")
        font.setPointSize(30)
        self.Info_Label_Authors.setFont(font)
        self.Info_Label_Authors.setStyleSheet("""QLabel {background: transparent; color: black;}""")
        self.verticalLayout_10.addWidget(self.Info_Label_Authors)
        self.Info_textBrowser_Authors = QtWidgets.QTextBrowser(parent=self.Info)
        self.Info_textBrowser_Authors.setObjectName("Info_textBrowser_Authors")
        self.Info_textBrowser_Authors.setStyleSheet("""
        QTextBrowser {
            background: transparent;
            border: none;
        }""")
        font.setFamily("Counter-Strike")
        font.setPointSize(20)
        self.Info_textBrowser_Authors.setFont(font)
        self.verticalLayout_10.addWidget(self.Info_textBrowser_Authors)
        self.verticalLayout_8.addLayout(self.verticalLayout_10)
        self.verticalLayout_9 = QtWidgets.QVBoxLayout()
        self.verticalLayout_9.setObjectName("verticalLayout_9")
        self.Info_Label_Programm = QtWidgets.QLabel(parent=self.Info)
        self.Info_Label_Programm.setObjectName("Info_Label_Programm")
        font.setFamily("Counter-Strike")
        font.setPointSize(30)
        self.Info_Label_Programm.setFont(font)
        self.Info_Label_Programm.setStyleSheet("""QLabel {background: transparent; color: black;}""")
        self.verticalLayout_9.addWidget(self.Info_Label_Programm)
        self.Info_textBrowser_Programm = QtWidgets.QTextBrowser(parent=self.Info)
        self.Info_textBrowser_Programm.setObjectName("Info_textBrowser_Programm")
        self.Info_textBrowser_Programm.setStyleSheet("""
        QTextBrowser {
            background: transparent;
            border: none;
            padding: 0;
        }""")
        self.verticalLayout_9.addWidget(self.Info_textBrowser_Programm)
        self.verticalLayout_8.addLayout(self.verticalLayout_9)
        self.horizontalLayout_7.addLayout(self.verticalLayout_8)

        self.Info_textBrowser_Additional = QtWidgets.QTextBrowser(parent=self.Info)
        self.Info_textBrowser_Additional.setObjectName("Info_textBrowser_Additional")
        self.Info_textBrowser_Additional.setStyleSheet("""
        QTextBrowser {
            background: transparent;
            border: none;
            font-size: 16pt;
            color: black;
        }""")
        self.verticalLayout_11 = QtWidgets.QVBoxLayout()
        self.verticalLayout_11.setObjectName("verticalLayout_11")
        self.verticalLayout_11.addWidget(self.Info_textBrowser_Additional)

        self.horizontalLayout_10 = QtWidgets.QHBoxLayout()
        self.horizontalLayout_10.setObjectName("horizontalLayout_10")
        spacerItem12 = QtWidgets.QSpacerItem(40, 20, QtWidgets.QSizePolicy.Policy.Expanding, QtWidgets.QSizePolicy.Policy.Minimum)
        self.horizontalLayout_10.addItem(spacerItem12)
        spacerItem13 = QtWidgets.QSpacerItem(40, 20, QtWidgets.QSizePolicy.Policy.Expanding, QtWidgets.QSizePolicy.Policy.Minimum)
        self.horizontalLayout_10.addItem(spacerItem13)
        self.verticalLayout_11.addLayout(self.horizontalLayout_10)
        self.horizontalLayout_7.addLayout(self.verticalLayout_11)
        self.verticalLayout_7.addLayout(self.horizontalLayout_7)
        self.horizontalLayout_8 = QtWidgets.QHBoxLayout()
        self.horizontalLayout_8.setObjectName("horizontalLayout_8")

        # --- Кнопка "НАЗАД" (возврат на предыдущую страницу) ---
        self.Info_pushButton_Back = QtWidgets.QPushButton(parent=self.Info)
        self.Info_pushButton_Back.setObjectName("Info_pushButton_Back")
        self.Info_pushButton_Back.clicked.connect(self.return_from_info)
        self.Info_pushButton_Back.setCursor(QtGui.QCursor(QtCore.Qt.CursorShape.CrossCursor))
        font = QtGui.QFont()
        font.setFamily("Counter-Strike")
        font.setPointSize(24)
        self.Info_pushButton_Back.setFont(font)
        self.Info_pushButton_Back.setStyleSheet("""
        QPushButton {background: transparent;
        color: black;  /* Черный цвет текста */
        border: none;  /* Убираем границу */
        padding: 0px;  /* Убираем отступы */}
        QPushButton:hover {color: #333;  /* Темно-серый при наведении */}QPushButton:pressed {color: #555;  /* Серый при нажатии */}""")
        self.horizontalLayout_8.addWidget(self.Info_pushButton_Back)
        spacerItem14 = QtWidgets.QSpacerItem(40, 20, QtWidgets.QSizePolicy.Policy.Expanding, QtWidgets.QSizePolicy.Policy.Minimum)
        self.horizontalLayout_8.addItem(spacerItem14)
        self.verticalLayout_7.addLayout(self.horizontalLayout_8)
        self.gridLayout_11.addLayout(self.verticalLayout_7, 0, 0, 1, 1)
        self.gridLayout_12.addLayout(self.gridLayout_11, 0, 0, 1, 1)
        self.stackedWidget.addWidget(self.Info)

        # ====================== НАСТРОЙКА ОТОБРАЖЕНИЯ ======================
        self.gridLayout.addWidget(self.stackedWidget, 0, 0, 1, 1)

        self.retranslateUi(mmCSGO)
        self.stackedWidget.setCurrentIndex(0)
        QtCore.QMetaObject.connectSlotsByName(mmCSGO)

        mmCSGO.resizeEvent = self.resizeEvent
        self.init_scalable_widgets()

    def init_scalable_widgets(self):

        self.scalable_widgets = [
        # Главное меню
        (self.Main_Label_mmCSGO, 0.12, 20),          # Главный заголовок
        (self.Main_pushButton_Start, 0.045, 12),     # Кнопка НАЧАТЬ
        (self.Main_pushButton_Info, 0.045, 12),       # Кнопка ИНФО
        (self.Main_pushButton_Exit, 0.045, 12),       # Кнопка ВЫХОД

        # Страница Plan (План задачи)
        (self.Plan_Label_Plan, 0.08, 16),            # Заголовок "ПЛАН ЗАДАЧИ"
        (self.Plan_Label_Parameters, 0.06, 14),       # Подзаголовок "Введите параметры:"
        (self.Plan_lineEdit_a, 0.04, 5),             # Поле ввода a:
        (self.Plan_lineEdit_p, 0.04, 5),             # Поле ввода P:
        (self.Plan_lineEdit_q, 0.04, 5),             # Поле ввода Q:
        (self.Plan_Label_Problem, 0.035, 9),          # Текст с ОДЗ
        (self.Plan_pushButton_Standart, 0.04, 10),    # Кнопка "Стандартные значения"
        (self.Plan_pushButton_Random, 0.04, 10),      # Кнопка "Сгенерировать значения"
        (self.Plan_pushButton_Info, 0.04, 10),        # Кнопка ИНФО
        (self.Plan_pushButton_Back, 0.04, 10),        # Кнопка НАЗАД
        (self.Plan_pushButton_Next, 0.04, 10),        # Кнопка ДАЛЕЕ

        # Страница Decision (Решение задачи)
        (self.Decision_Label_Decision, 0.08, 16),     # Заголовок "РЕШЕНИЕ ЗАДАЧИ"
        (self.Decision_Label_Parameters, 0.06, 14),   # Подзаголовок "Введенные параметры:"
        (self.Decision_Label_a, 0.05, 12),           # Метка a:
        (self.Decision_Label_p, 0.05, 12),           # Метка P:
        (self.Decision_Label_q, 0.05, 12),           # Метка Q:
        (self.Decision_Label_Task, 0.06, 14),        # Подзаголовок "Условие задачи:"
        (self.Desion_Label, 0.05, 12),               # Метка Ответ:
        (self.Decision_pushButton_Back, 0.04, 10),   # Кнопка НАЗАД
        (self.Decision_pushButton_Exit, 0.04, 10),    # Кнопка ВЫХОД
        (self.Decision_pushButton_Info, 0.04, 10),    # Кнопка ИНФО

        # Страница Info (Справочник)
        (self.Info_Label_Info, 0.08, 16),            # Заголовок "СПРАВОЧНИК mmCS:GO"
        (self.Info_Label_Authors, 0.06, 14),         # Подзаголовок "АВТОРЫ:"
        (self.Info_Label_Programm, 0.06, 14),        # Подзаголовок "О ПРОГРАММЕ:"
        (self.Info_pushButton_Back, 0.04, 10),       # Кнопка НАЗАД
        (self.Decision_textBrowser_Task, 0.035, 9),
        (self.Info_textBrowser_Authors, 0.035, 9),
        (self.Info_textBrowser_Programm, 0.035, 9),
        (self.Info_textBrowser_Additional, 0.035, 9)
    ]

        # Настройка QTextBrowser с относительными размерами
        for browser in [self.Decision_textBrowser_Task,
                       self.Info_textBrowser_Authors,
                       self.Info_textBrowser_Programm]:
            browser.setStyleSheet("""
                QTextBrowser {
                    font-size: 1.8vh;
                    background: transparent;
                    border: none;
                }""")

    def resizeEvent(self, event):
        """Обработчик изменения размера окна"""
        for widget, ratio, min_size in self.scalable_widgets:
            if widget and hasattr(widget, 'setFont'):
                new_size = max(min_size, int(self.stackedWidget.height() * ratio))
                font = widget.font()
                font.setPointSize(new_size)
                widget.setFont(font)

        super(type(self.stackedWidget.parent()), self.stackedWidget.parent()).resizeEvent(event)

    # Функция отправления на Info
    def go_to_info(self, from_page):
        self.previous_page = from_page
        self.stackedWidget.setCurrentIndex(3)

    # Функция возврата из Info
    def return_from_info(self):
        self.stackedWidget.setCurrentIndex(self.previous_page)

    def create_plot_widget(self):
        """Создает виджет с графиком и возвращает его"""
        figure = Figure(figsize=(6, 6), facecolor='none')
        canvas = FigureCanvas(figure)
        canvas.setStyleSheet("background-color: transparent;")
        canvas.setMinimumSize(300, 300)

        # Отрисовка графика
        ax = figure.add_subplot(111)
        ax.set_aspect('equal')
        ax.set_axis_off()

        lw = 1.6

        # Основные линии
        ax.plot([0, 6], [0, 0], color='black', lw=lw)
        ax.plot([0, 5], [0, 2], color='black', lw=lw)

        # Штриховка основания
        for i in range(30):
            ax.plot([i * 0.2, i * 0.2 - 0.15], [0, -0.2], color='black', lw=lw * 0.5)

        # Нижняя окружность
        r = 0.3
        cx = 5 / 3
        cy = (2 / 5) * cx + r
        circle = plt.Circle((cx, cy), r, color='white', ec='black', lw=lw)
        ax.add_artist(circle)
        ax.plot(cx, cy, 'ko', markersize=2)

        # Угол a
        ax.text(0.6, -0.05 + 0.35 * r, 'a', fontsize=12)

        # Линии из центра нижней окружности
        l = np.sqrt((5 - cx) ** 2 + (2 - cy) ** 2)
        l /= 6
        x1 = cx + l
        y1 = cy + l * (2 / 5)
        ax.plot([cx, x1], [cy, y1], color='black', lw=lw)

        angle = 45
        radians = np.radians(angle)
        x2 = cx + np.cos(radians) * 10 * r
        y2 = cy + np.sin(radians) * 10 * r
        ax.plot([cx, x2], [cy, y2], color='black', lw=lw)

        # Угол b
        ax.text(cx + 0.32, cy + 0.18, 'b', fontsize=10)

        # Верхняя окружность
        ex = x2
        ey = y2
        cx2 = ex + np.cos(radians) * r
        cy2 = ey + np.sin(radians) * r - r * 1.4
        circle2 = plt.Circle((cx2, cy2), r, color='white', ec='black', lw=lw)
        ax.add_artist(circle2)
        ax.plot(cx2, cy2, 'ko', markersize=2)

        # Вертикальная линия вниз от верхней окружности
        sx = cx2 + r
        sy = cy2
        l2 = 1.5 * r * 4 / 3
        ey2 = sy - l2
        ax.plot([sx, sx], [sy, ey2], color='black', lw=lw)

        # Прямоугольник
        s = 0.2
        sx2 = sx - s / 2
        sy2 = ey2 - s
        square = plt.Rectangle((sx2, sy2), s, s, fill=False, ec='black', lw=lw)
        ax.add_artist(square)

        # Центр квадрата и стрелка вниз
        cx3 = sx2 + s / 2
        cy3 = sy2 + s / 2
        ax.plot(cx3, cy3, 'ko', markersize=2)
        ax.arrow(cx3, cy3, 0, -0.3, head_width=0.05, head_length=0.05, fc='black', ec='black')

        # Верхний треугольник
        top = (cx2, cy2)
        tw = 0.25 * 2 / 3
        th = 0.4
        bl = (cx2 - tw, cy2 + th)
        br = (cx2 + tw, cy2 + th)
        ax.plot([bl[0], top[0], br[0]], [bl[1], top[1], br[1]], color='black', lw=lw)

        # Прямоугольник со штриховкой
        rh = 0.25
        ew = 0.15
        rl = bl[0] - ew
        rr = br[0] + ew
        rb = bl[1]
        rt = rb + rh

        ax.plot([rl, rr], [rb, rb], color='black', lw=lw)
        ax.plot([rl, rr], [rt, rt], color='black', lw=lw)
        ax.plot([rl, rl], [rb, rt], color='black', lw=lw)
        ax.plot([rr, rr], [rb, rt], color='black', lw=lw)

        # Штриховка
        sp = 0.25 / 5
        x_start = rl
        while x_start - rh <= rr:
            x1 = x_start
            y1 = rt
            x2 = x_start - rh
            y2 = rb

            if x2 < rl:
                dx = rl - x2
                dy = dx
                x2 = rl
                y2 = rb + dy

            if x1 > rr:
                dx = x1 - rr
                dy = dx
                x1 = rr
                y1 = rt - dy

            ax.plot([x1, x2], [y1, y2], color='black', lw=lw * 0.5)
            x_start += sp

        # Стрелка вниз из нижней окружности
        ax.arrow(cx, cy, 0, -2.5 * r, head_width=0.05, head_length=0.05, fc='black', ec='black')

        # Подписи
        ax.text(cx - 0.35 - 0.6 * r, cy + 0.02, 'B', fontsize=12)
        ax.text(cx - 0.2, cy - 2.5 * r - 0.04, 'P', fontsize=12)
        ax.text(cx2 - 0.4 - 0.6 * r, cy2 + 0.08, 'C', fontsize=12)
        ax.text(cx3 + 0.05, cy3 - 0.3, 'Q', fontsize=12)

        canvas.draw()
        return canvas

    def retranslateUi(self, mmCSGO):
        _translate = QtCore.QCoreApplication.translate
        mmCSGO.setWindowTitle(_translate("mmCSGO", "mmCS:GO"))
        self.Main_pushButton_Info.setText(_translate("mmCSGO", "ИНФО"))
        self.Main_pushButton_Start.setText(_translate("mmCSGO", "НАЧАТЬ"))
        self.Main_pushButton_Exit.setText(_translate("mmCSGO", "ВЫХОД"))
        self.Main_Label_mmCSGO.setText(_translate("mmCSGO", "mmCS:GO"))
        self.Plan_Label_Plan.setText(_translate("mmCSGO", "ПЛАН ЗАДАЧИ"))
        self.Plan_Label_Parameters.setText(_translate("mmCSGO", "Введите параметры:"))
        self.Plan_lineEdit_a.setText(_translate("mmCSGO", "a:"))
        self.Plan_lineEdit_p.setText(_translate("mmCSGO", "P:"))
        self.Plan_lineEdit_q.setText(_translate("mmCSGO", "Q:"))
        self.Plan_Label_Problem.setText(_translate("mmCSGO", "Значения параметра a допускаются в пределах от 0 до 90."))
        self.Plan_pushButton_Standart.setText(_translate("mmCSGO", "Стандартные значения"))
        self.Plan_pushButton_Random.setText(_translate("mmCSGO", "Сгенерировать значения"))
        self.Plan_pushButton_Info.setText(_translate("mmCSGO", "ИНФО"))
        self.Plan_pushButton_Back.setText(_translate("mmCSGO", "НАЗАД"))
        self.Plan_pushButton_Next.setText(_translate("mmCSGO", "ДАЛЕЕ"))
        self.Decision_Label_Decision.setText(_translate("mmCSGO", "РЕШЕНИЕ ЗАДАЧИ"))
        self.Decision_Label_Parameters.setText(_translate("mmCSGO", "Введенные параметры:"))
        self.Decision_Label_a.setText(_translate("mmCSGO", "a:"))
        self.Decision_Label_p.setText(_translate("mmCSGO", "P:"))
        self.Decision_Label_q.setText(_translate("mmCSGO", "Q:"))
        self.Decision_Label_Task.setText(_translate("mmCSGO", "Условие задачи:"))
        self.Decision_textBrowser_Task.setHtml(_translate("mmCSGO", """<!DOCTYPE HTML PUBLIC "-//W3C//DTD HTML 4.0//EN" "http://www.w3.org/TR/REC-html40/strict.dtd">
<html>
<head>
    <meta name="qrichtext" content="1" />
    <style type="text/css">
        body {
            font-family: 'Counter-Strike';
            font-size: 16pt;
            color: black;
            background-color: transparent;
        }
        p {
            margin: 6px 0;
            -qt-block-indent: 0;
            text-indent: 0px;
        }
    </style>
</head>
<body>
    <p align="justify">Два груза, веса которых равны P и Q, действуют на нить, перекинутую через блок C. Груз Q висит свободно, а груз P опирается на плоскость, образующую с горизонтом угол a; вся система находится в равновессии. Найдите угол B, который образует прямая BC с наклонной плоскостью, и давление N груза P на плоскость.</p>
</body>
</html>"""))
        self.Desion_Label.setText(_translate("mmCSGO", "Ответ:"))
        self.Decision_pushButton_Back.setText(_translate("mmCSGO", "НАЗАД"))
        self.Decision_pushButton_Exit.setText(_translate("mmCSGO", "ВЫХОД"))
        self.Decision_pushButton_Info.setText(_translate("mmCSGO", "ИНФО"))
        self.Info_Label_Info.setText(_translate("mmCSGO", "СПРАВОЧНИК mmCS:GO"))
        self.Info_Label_Authors.setText(_translate("mmCSGO", "<html><head/><body><p align=\"center\">АВТОРЫ:</p></body></html>"))
        self.Info_textBrowser_Authors.setHtml(_translate("mmCSGO", """<!DOCTYPE HTML PUBLIC "-//W3C//DTD HTML 4.0//EN" "http://www.w3.org/TR/REC-html40/strict.dtd">
<html>
<head>
    <meta name="qrichtext" content="1" />
    <style type="text/css">
        body {
            font-family: 'Counter-Strike';
            font-size: 20pt;
            color: black;
            background-color: transparent;
        }
        p {
            margin: 0;
            -qt-block-indent: 0;
            text-indent: 0px;
        }
    </style>
</head>
<body>
    <p align="center">Кузнецов А.А.</p>
    <p align="center">Свинтаржицкий Я.В.</p>
    <p align="center">Хамзатов Р.А.</p>
</body>
</html>"""))
        self.Info_Label_Programm.setText(_translate("mmCSGO", "<html><head/><body><p align=\"center\">О ПРОГРАММЕ:</p></body></html>"))
        self.Info_textBrowser_Programm.setHtml(_translate("mmCSGO", """<!DOCTYPE HTML PUBLIC "-//W3C//DTD HTML 4.0//EN" "http://www.w3.org/TR/REC-html40/strict.dtd">
<html>
<head>
    <meta name="qrichtext" content="1" />
    <style type="text/css">
        body {
            font-family: 'Counter-Strike';
            font-size: 18pt;
            color: black;
            background-color: transparent;
        }
        p {
            margin: 0;
            -qt-block-indent: 0;
            text-indent: 0px;
        }
    </style>
</head>
<body>
    <p align="justify">Программа "mmCS:GO" предназначена для решения задачи по механике, предоставленной группе #7. Программа позволяет пользователю ввести в задачу свои параметры или сгенерировать случайные.</p>
</body>
</html>"""))
        self.Info_textBrowser_Additional.setHtml(_translate("mmCSGO", """<!DOCTYPE HTML PUBLIC "-//W3C//DTD HTML 4.0//EN" "http://www.w3.org/TR/REC-html40/strict.dtd">
<html>
<head>
    <meta name="qrichtext" content="1" />
    <style type="text/css">
        body {
            font-size: 18pt;
            color: black;
            background-color: transparent;
            line-height: 1.5;
        }
        p {
            margin: 10px 0;
            -qt-block-indent: 0;
            text-indent: 0px;
        }
        .math {
            font-size: 20px;
            display: inline-block;
            vertical-align: middle;
            margin: 2px 0;
        }
        .math mfrac {
            font-size: 18px;
        }
        .section {
            margin-bottom: 15px;
        }
    </style>
</head>
<body>

<p><p style="font-weight: bold;">Разложим P на компоненты:</p><p style="text-indent: 1.5em;"><math xmlns="http://www.w3.org/1998/Math/MathML" style="font-size:20px; "><mrow><msub><mi>P</mi><mi>y</mi></msub><mo>=</mo><mi>P</mi><mi>cos</mi><mi>&alpha;</mi></mrow><mrow><msub><mi>, P</mi><mi>x</mi></msub><mo>=</mo><mi>P</mi><mi>sin</mi><mi>&alpha;</mi></mrow></math></p><p style="font-weight: bold;">Запишем услвия равновесия для обоих грузов:</p><p style="font-weight: bold;">Для P:</p><p style="text-indent: 1.5em;"><math xmlns="http://www.w3.org/1998/Math/MathML" style="font-size:20px"><mrow><msub><mi>O</mi><mi>x</mi></msub><mo>:</mo><mi>T</mi><mi>cos</mi><mi>&theta;</mi><mo>=</mo><mi>P</mi><mi>sin&theta;</mi><mo>(1)</mo></mrow></math></p><p style="text-indent: 1.5em;"><math xmlns="http://www.w3.org/1998/Math/MathML" style="font-size:20px"><mrow><msub><mi>O</mi><mi>y</mi></msub><mo>:</mo><mi>N</mi><mo>-</mo><mi>P</mi><mi>cos</mi><mi>&alpha;</mi><mo>-</mo><mi>T</mi><mi>sin</mi><mi>&theta;</mi><mo>=</mo><mn>0</mn></mrow></math></p><p style="font-weight: bold;">Для Q:</p><p style="text-indent: 1.5em;"><math xmlns="http://www.w3.org/1998/Math/MathML" style="font-size:20px"><mrow><mi>T</mi><mo>=</mo><mi>Q</mi><mo>(2)</mo></mrow></math></p><p style="font-weight: bold;">Отсюда:</p><p style="text-indent: 1.5em;"><math xmlns="http://www.w3.org/1998/Math/MathML" style="font-size:20px"><mrow><mi>N</mi><mo>=</mo><mi>P</mi><mi>cos</mi><mi>&alpha;</mi><mo>+</mo><mi>T</mi><mi>sin</mi><mi>&theta;</mi><mo>(3)</mo></mrow></math></p><p style="font-weight: bold;">Подставим (2) в (1):</p><p style="text-indent: 1.5em;"><math xmlns="http://www.w3.org/1998/Math/MathML" style="font-size:20px"><mrow><mi>Q</mi><mi>cos</mi><mi>&theta;</mi><mo>=</mo><mi>P</mi><mi>sin</mi><mi>&alpha;</mi></mrow></math></p><p style="text-indent: 1.5em;"><math xmlns="http://www.w3.org/1998/Math/MathML" style="font-size:20px"><mi>cos</mi><mi>&theta;</mi><mo>=</mo><mfrac><mrow><mi>P</mi><mi>sin</mi><mi>&alpha;</mi></mrow><mrow><mi>Q</mi></mrow></mfrac></math></p><p style="font-weight: bold;">Подставим (2) в (3):</p><p style="text-indent: 1.5em;"><math xmlns="http://www.w3.org/1998/Math/MathML" style="font-size:20px"><mrow><mi>N</mi><mo>=</mo><mi>P</mi><mi>cos</mi><mi>&alpha;</mi><mo>+</mo><mi>Q</mi><mi>sin</mi><mi>&theta;</mi></mrow></math></p><p style="text-indent: 1.5em;"><math xmlns="http://www.w3.org/1998/Math/MathML" style="font-size:20px"><mrow><mi>sin</mi><mi>&theta;</mi><mo>=</mo><msqrt><mn>1</mn><mo>-</mo><msup><mi>cos</mi><mn>2</mn></msup><mi>&theta;</mi></msqrt><mo>=</mo><msqrt><mn>1</mn><mo>-</mo><mo></mo><mfrac><mrow><msup><mi>P</mi><mn>2</mn></msup><msup><mi>sin</mi><mn>2</mn></msup><mi>&alpha;</mi></mrow><mrow><msup><mi>Q</mi><mn>2</mn></msup></mrow></mfrac></msqrt></mrow></math></p><p style="text-indent: 1.5em;"><math xmlns="http://www.w3.org/1998/Math/MathML" style="font-size:20px"><mrow><mi>N</mi><mo>=</mo><mi>P</mi><mi>cos</mi><mi>&alpha;</mi><mo>+</mo><mi>Q</mi><msqrt><mn>1</mn><mo>-</mo><mo></mo><mfrac><mrow><msup><mi>P</mi><mn>2</mn></msup><msup><mi>sin</mi><mn>2</mn></msup><mi>&alpha;</mi></mrow><mrow><msup><mi>Q</mi><mn>2</mn></msup></mrow></mfrac></msqrt><mo>=</mo><mi>P</mi><mi>cos</mi><mi>&alpha;</mi><mo>+</mo><msqrt><msup><mi>Q</mi><mn>2</mn></msup><mo>-</mo><msup><mi>P</mi><mn>2</mn></msup><msup><mi>sin</mi><mn>2</mn></msup><mi>&alpha;</mi></msqrt></mrow></math></p><p style="font-weight: bold;">Ответ:</p><p style="text-indent: 1.5em;"><math xmlns="http://www.w3.org/1998/Math/MathML" style="font-size:20px"><mi>cos</mi><mi>&theta;</mi><mo>=</mo><mfrac><mrow><mi>P</mi><mi>sin</mi><mi>&alpha;</mi></mrow><mrow><mi>Q</mi></mrow></mfrac></math></p><p style="text-indent: 1.5em;"><math xmlns="http://www.w3.org/1998/Math/MathML" style="font-size:20px"><mrow><mi>N</mi><mo>=</mo><mi>P</mi><mi>cos</mi><mi>&alpha;</mi><mo>+</mo><msqrt><msup><mi>Q</mi><mn>2</mn></msup><mo>-</mo><msup><mi>P</mi><mn>2</mn></msup><msup><mi>sin</mi><mn>2</mn></msup><mi>&alpha;</mi></msqrt></mrow></math></p></p>
</body>
</html>"""))
        self.Info_pushButton_Back.setText(_translate("mmCSGO", "НАЗАД"))

if __name__ == "__main__":
    import sys
    app = QtWidgets.QApplication(sys.argv)
    mmCSGO = QtWidgets.QWidget()
    ui = Ui_mmCSGO()
    ui.setupUi(mmCSGO)
    mmCSGO.show()
    sys.exit(app.exec())
```
